### 리스트
* 리스트는 변수들을 한 줄로 이어 붙여놓은 개념
* 만약 변수 a, b, c가 있을 때, 리스트는 [a,b,c]
* 리스트의 변수는 0번부터 시작, 즉 list = [a,b,c]라고 정의할 때, list[0] = a, list[1] = b, list[2] = c

```python
a = []
b = [1,2,3]
c = [1, 2, 'Hello', 4]
d = [1, 2, 3, ['a','b','c']]
```

* 리스트 안의 성분들의 index
    - 0번부터 시작. 즉 b의 첫 번째 성분은 b[0]
    - 마지막 성분의 index는 -1, 즉 b의 마지막 성분은 b[-1]
    - 리스트의 성분으로 또 다른 리스트가 저장된 이중리스트의 성분들은 이중 index를 사용하여 표현

* 리스트의 특정 성분들 골라내기(slicing)
    - :(콜론)을 기준으로 왼쪽은 시작지점의 다음 값(0부터 시작), 오른쪽은 끝나는 지점
    - 즉, c[1:3] 이면 리스트 c의 2번째 성분부터 3번째 성분까지
    - : 왼쪽 값을 지정하지 않으면 처음부터. 즉 c[:2]는 처음부터 2번째 성분까지
    - : 오른쪽 값을 지정하지 않으면 마지막까지, c[1:]은 2번째 성분부터 마지막까지

* 리스트 연산
    - ' + ' 기호는 성분들의 더하기가 아니라 리스트를 이어붙여서 새로운 하나의 리스트를 만들게됨. 따라서 '리스트 + 리스트' 형태
    - ' * ' 기호는 리스트를 반복. 따라서 ' 리스트 * 정수 ' 형태로 작성

```python
b + c
```

```python
b + [3]
```

```python
b * 2
```

* 리스트의 수정과 삭제
     - 리스트의 성분을 다른 값으로 대체하여 수정
     - del 함수를 이용하여 리스트 성분 삭제

```python
a = [1, 2, 3, 4]
print(a)
a[1] = 10
print(a)
```

```python
del a[1]
print(a)
```

```python
del a[1:]
print(a)
```

```python
len(a)
```

```python
del a
a
```

예제1. 반지름을 1부터 50까지 중 홀수 25개(1, 3, 5, 7, ...)으로 가지는 원의 넓이를 출력하여라.

예제2. for 를 이용한 20번의 반복을 통해 리스트의 성분을 증가시키며 리스트를 업데이트하는데, 매번의 반복마다 앞의 두 항의 합을 새로운 성분으로 가지는(피보나치 수열) 리스트를 출력하여라. 즉, 처음 리스트는 [1,1], 한 번 반복한 후 얻어지는 리스트는 [1,1,2], 다음 반복에서 얻어지는 리스트는 [1,1,2,3], 그 다음은 [1,1,2,3,5], ...

```python
a = [1,1]
for l in range(1,21):
    a.append(a[-2]+a[-1])
    print('iteration number', l, ', Fibonacci sequence : ', a)
```

```python
for i in a.keys():
    print(i)
```

```python
a.values()
```

```python
for i in a.values():
    print(i)
```

```python
a.items()
```

```python
a.clear()
a
```

# library/module 사용

* 모듈(module)
    - 모듈은 파이썬 코드 파일(.py) 형태로 저장되어있는 함수나 변수 또는 클래스를 모아 놓은 파이썬 파일
    - 다른 사람들이 이미 만들어 놓은 모듈을 사용할 수도 있고 직접 만들어 사용 가능
    - 많은 경우, 다른 사람들이 이미 만들어 놓은 모듈을 불러와서 필요한 함수들을 사용
* 라이브러리(library), 패키지(package)
    - 여러 모듈을 모아놓은 '도서관'
    - 많이 사용되는 라이브러리들은 파이썬을 설치할 때 자동적으로 설치됨

* import module
    - 파이썬에서는 저장되어있는 모듈을 불러와서 모듈안에 들어있는 함수들을 사용할 수 있다
    - 모듈안에는 수많은 함수들이 저장되어있는 경우들이 있기 때문에 필요한 함수만 불러와서 사용할 수도 있다
    - 불러온 모듈의 이름이 module이고 module 모듈안에있는 ftn이라는 함수를 사용하기 위해서는 아래와 같이 .을 사용하여 함수를 사용
    > import module  
      module.ftn()

```python
sqrt(4)
```

```python
import math
math.sqrt(4)
```

   * 모듈안에 저장되어있는 함수를 확인하기 위해서는 dir(module) 또는 module.+tab

```python
dir(math)
```

* 많은 함수들을 다 불러오기 보다 특정함수만 불러와서 사용하고 싶을 때는
   > from module import ftn
* 이 경우에는 module.ftn이 아니라 그냥 ftn만 사용해서 함수 표현

```python
log(2)
```

```python
del math
from math import log
math.log(2)
```

   * 모듈안의 모든 함수, 변수, 클래스를 가지고 오면 module.을 안붙여도 가능
   > from module import *
   * import module 과는 다름

```python
from math import *
print(log(12))
exp(0)
```

   * 모듈의 이름이 긴 경우 파이썬을 사용하는 동안 다른 이름을 지정해서 사용할 수 있음.
       - 예를 들어 module라는 모듈의 이름을 파이썬을 사용하는 동안 m이라는 이름으로 간편하게 지정
   > import module as m

```python
import math as m
m.sqrt(4)
```

### Pandas Series
판다스 시리즈Pandas Series는 1차원 array형태이며, 데이터 테이블에서의 열을 표현.

```python
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
print(myvar[0])
```

```python
a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

```python
print(myvar[0]); print(myvar[1]); print(myvar[2])
```

```python
print(myvar['x']);print(myvar['y']);print(myvar['z'])
```
